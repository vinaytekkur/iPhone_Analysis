```python
import pandas as pd
```

```python
# 1 - The column names have spaces . rename the column names to have underscore '_' instead of space 
# (try to do in one go instead of specifying each column nam in rename method)
```

```python
df_iphones = pd.read_csv('iphone.csv')
```

```python
df_iphones.columns = df_iphones.columns.str.replace(' ','_')
```

```python
df_iphones
```

```python
################################################################################################
```

```python
# 2- start rating for some of the models is missing in the dataset. 
# fill those missing values with the average rating all the models.
```

```python
avg_rating = round(df_iphones['Star_Rating'].mean(),2)
```

```python
df_iphones['Star_Rating'].fillna(value=avg_rating, inplace=True)
```

```python
################################################################################################
```

```python
# 3- Now instead of filling missing values with avg rating of full dataset , 
# fill with avg rating based on RAM. example : 
# if rating for a 2 gb phone is missing then take average of all other 2 gb phones rating and fill that value. 
```

```python
df_phones_ram_avg_rating = df_iphones.groupby('Ram', as_index=False).agg({'Star_Rating':'mean'})
```

```python
df_iphones_merged = pd.merge(left=df_iphones, right=df_phones_ram_avg_rating, how='left', left_on='Ram', right_on='Ram')
```

```python
df_iphones_merged
```

```python
df_iphones_merged['Star_Rating_x'].fillna(value=df_iphones_merged['Star_Rating_y'], inplace=True)
```

```python
################################################################################################
```

```python
# 4- create a new column in the dataframe "Discount_Percentage" based on MRP and sale value
```

```python
df_iphones['Discount_Percentage'] = round(((df_iphones['Mrp'] - df_iphones['Sale_Price']) / df_iphones['Mrp']) * 100)
```

```python
# 11 - find total no of reviews for iphone 11 and iphone 12 category . Output should have only 2 rows (for 11 and 12).
```

```python
iphone_11_tot_reviews = [df_iphones[df_iphones['Model_Name'].str.contains('11')]['Number_Of_Ratings'].sum()]
iphone_12_tot_reviews = [df_iphones[df_iphones['Model_Name'].str.contains('12')]['Number_Of_Ratings'].sum()]
```

```python
iphone11_df = pd.DataFrame(iphone_11_tot_reviews, index=['iphone_11'], columns=['Total_Reviews'])
iphone12_df = pd.DataFrame(iphone_12_tot_reviews, index=['iphone_12'], columns=['Total_Reviews'])
```

```python
pd.concat([iphone11_df, iphone12_df])
```

```python
# 12- which iphone has 3rd highest MRP
```

```python
# 14- which iphone with 128 GB space has highest ratings to review ratio
```

```python
iphones_with_128_storage = df_iphones[df_iphones['Product_Name'].str.contains('128')]
```

```python
iphones_with_128_storage
```

```python
iphones_with_128_storage.loc[iphones_with_128_storage.index,'rating_reviews_ratio'] =
round(iphones_with_128_storage.loc[iphones_with_128_storage.index,'Number_Of_Ratings'] / 
      iphones_with_128_storage.loc[iphones_with_128_storage.index,'Number_Of_Reviews'],2)
```

```python
iphones_with_128_storage.loc[iphones_with_128_storage.index,'ranking'] = 
iphones_with_128_storage['rating_reviews_ratio'].rank(method='dense', ascending=False)
```

```python
iphones_with_128_storage[iphones_with_128_storage['ranking'] == 1]
```

```python
# Validation
iphones_with_128_storage['rating_reviews_ratio'].max()
```
